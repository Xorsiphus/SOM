```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# data = pd.read_json("generated.json")
# data = pd.read_csv("ratings.json")
data = pd.read_csv("anime.csv")

# print(data.info())

data['genre'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(20, 20))
```

```python
lines = data['genre'].str.split(', ')
genres = {}
indexes = []

for line in lines:
    if type(line) is list:
        for genre in line:
            if genre not in genres:
                genres[genre] = 0
            else:
                genres[genre] += 1
# print(genres)

genres_data = {'genres': []}
counter = 0
for genre in genres:
    indexes.append(genre)
    genres_data['genres'].append(genres[genre])

# print(indexes)
# print(genres_data)

df = pd.DataFrame(genres_data, index=indexes)
df.plot.pie(autopct='%1.1f%%', figsize=(20, 20), y='genres')
```
